```python
import pandas as pd
import glob

# Specify the path to your CSV files
path = 'D:\Super_resolution_output_folder\*.csv'

# Use glob to get a list of all CSV files in the specified path
files = glob.glob(path)

# Initialize an empty DataFrame to store the combined data
combined_data = pd.DataFrame()

# Iterate through each CSV file and append its data to the combined_data DataFrame
for file in files:
    df = pd.read_csv(file)
    combined_data = combined_data.append(df, ignore_index=True)

# Save the combined data to a new CSV file
combined_data.to_csv(r'D:\Super_resolution_output_folder\combined_file.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Replace with the  path to your CSV file
file_path = r'D:\Super_resolution_output_folder\combined_file.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Extract the "Epoch" and "Average_PSNR" columns
epoch = df['Epoch']
average_psnr = df['Average_PSNR']

# Plot the data
plt.plot(epoch, average_psnr, marker='o', linestyle='-', color='b', label='Average PSNR')
#plt.title('Average PSNR Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('PSNR')
#plt.legend()
plt.grid(True)
plt.savefig('PSNR_plot.png')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#Repalce with your experimental results
models = ['LR', 'SR', 'HR']
values = {
    'Precision': [0.82,0.81,0.89],
    'Recall': [0.84,0.84,0.89],
    'Accuracy': [0.82,0.82,0.89],
    'F1 Score': [0.83,0.82,0.89]
}

# Plotting
bar_width = 0.2
index = np.arange(len(models))

fig, ax = plt.subplots(figsize=(10, 6))  # Adjust the figure size as needed

for i, metric in enumerate(values.keys()):
    ax.bar(index + i * bar_width, values[metric], bar_width, label=metric)

ax.set_xlabel('Models')
ax.set_ylabel('Performance Metrics')
ax.set_xticks(index + bar_width)
ax.set_xticklabels(models)

# Adjust legend position
ax.legend(loc='upper left', bbox_to_anchor=(1, 1))

# Save the plot with tight layout
plt.tight_layout()
plt.savefig('performance_plot.png', bbox_inches='tight')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Replace with the path to your CSV file
file_path = r'D:\cross_validation_5_epcoch_SR_results.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Extract the "Epoch", "Val_accuracy" and ["Accuracy"] columns
epoch = df['Epoch']
average_psnr = df['Val_accuracy']
another_line_data = df['Accuracy']

# Plot the data
plt.plot(epoch, average_psnr, marker='o', linestyle='-', color='purple', label='Validation Accuracy')
plt.plot(epoch, another_line_data, marker='o', linestyle='-', color='orange', label='Train Accuracy')

plt.xlabel('Epoch')
plt.ylabel('Accuracy')  
plt.legend()
plt.grid(True)
plt.savefig('combined_plot.png')
plt.show()
```
